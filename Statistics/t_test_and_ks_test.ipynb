# KS Test
*  T Test  Test  Varience  distribution  normal distribution
* : variance , t-test 

```python
# Use T-test
import numpy as np

group_a = np.random.normal(loc=0, scale=1.0, size=20)
group_b = np.random.normal(loc=0, scale=3.0, size=20)

print('group a: \n{}'.format(np.sort(group_a)))
print('group b: \n{}'.format(np.sort(group_b)))

stats.ttest_ind(group_a, group_b)
```

```python
# Use T-test
import numpy as np

group_a = np.random.normal(loc=0, scale=1.0, size=20)
group_b = np.random.normal(loc=0, scale=3.0, size=20)

print('group a: \n{}'.format(np.sort(group_a)))
print('group b: \n{}'.format(np.sort(group_b)))

stats.ttest_ind(group_a, group_b)
```

```python
stats.kstest(group_a, group_b)
```

```python
#samples from lognormal
lognorm_a = np.random.lognormal(mean=1, sigma=10, size=20)
print('group a: \n{}'.format(np.sort(lognorm_a)))

lognorm_b = np.random.lognormal(mean=1, sigma=10, size=20)
print('group a: \n{}'.format(np.sort(lognorm_b)))

stats.ttest_ind(lognorm_a, lognorm_b)
```

```python
stats.kstest(lognorm_a, lognorm_b)
```
