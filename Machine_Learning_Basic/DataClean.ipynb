```python
import pandas as pd

url="https://raw.githubusercontent.com/jbrownlee/Datasets/master/oil-spill.csv"
pd_csv=pd.read_csv(url,header=None)
```

```python
pd_csv.head(10)
```

```python
import numpy as np
count_unique = pd_csv.nunique()
```

## Count the number of unique values
* delete those column with low number of the unique values

```python
count_unique
```

```python
drop_list = [ i for i, v in enumerate(count_unique) if v<=2]
drop_list
```

```python
pd_csv.drop(drop_list, axis=1)
```
